# cycle detect

```python
def dfs(graph):
    color = {}
    discovery = {}
    finish = {}
    predecessor = {}
    time = 0
    cycle_found = [False]  # Use a mutable object to track if a cycle is found

    # Initialize all nodes
    for u in graph:
        color[u] = 'WHITE'
        predecessor[u] = None

    # Perform DFS
    for u in graph:
        if color[u] == 'WHITE':
            time = dfs_visit(graph, u, color, discovery, finish, predecessor, time, cycle_found)

    return discovery, finish, predecessor, cycle_found[0]


def dfs_visit(graph, u, color, discovery, finish, predecessor, time, cycle_found):
    color[u] = 'GRAY'
    time += 1
    discovery[u] = time

    for v in graph[u]:
        if color[v] == 'WHITE':
            predecessor[v] = u
            time = dfs_visit(graph, v, color, discovery, finish, predecessor, time, cycle_found)
        elif color[v] == 'GRAY':  # Back edge detected, indicating a cycle
            cycle_found[0] = True  ##############
            print("Cycle found!")  ############

    color[u] = 'BLACK'
    time += 1
    finish[u] = time
    return time


# Example graph with a cycle
graph = {
    'A': ['B'],
    'B': ['C'],
    'C': ['A'],  # Cycle: A -> B -> C -> A
    'D': ['E'],
    'E': []
}

discovery, finish, predecessor, cycle_detected = dfs(graph)
print("Discovery Times:", discovery)
print("Finish Times:", finish)
print("Predecessors:", predecessor)
print("Cycle Detected:", cycle_detected)
```

# undirected

```python
def dfs(graph):
    color = {}
    discovery = {}
    finish = {}
    predecessor = {}
    time = 0
    cycle_found = [False]  # Use a mutable object to track if a cycle is found

    # Initialize all nodes
    for u in graph:
        color[u] = 'WHITE'
        predecessor[u] = None

    # Perform DFS
    for u in graph:
        if color[u] == 'WHITE':
            time = dfs_visit(graph, u, color, discovery, finish, predecessor, time, cycle_found)

    return discovery, finish, predecessor, cycle_found[0]


def dfs_visit(graph, u, color, discovery, finish, predecessor, time, cycle_found):
    color[u] = 'GRAY'
    time += 1
    discovery[u] = time

    for v in graph[u]:
        if color[v] == 'WHITE':
            predecessor[v] = u
            time = dfs_visit(graph, v, color, discovery, finish, predecessor, time, cycle_found)
        elif color[v] == 'GRAY' and predecessor[u] != v:  # Back edge detected in undirected graph #########################
            cycle_found[0] = True
            print("Cycle found!")

    color[u] = 'BLACK'
    time += 1
    finish[u] = time
    return time


# Cycle Detection:

# A cycle is detected if a GRAY node is encountered that is not the parent of the current node (predecessor[u] != v).
# This ensures that the edge being checked is not the one that led to the current node, which is expected in an undirected graph.
# Graph Representation:

# The graph is represented as an adjacency list, where each node points to its neighbors. For an undirected graph, each edge is bidirectional (e.g., if A is connected to B, then B is also connected to A).




# Example undirected graph with a cycle
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D'],
    'C': ['A', 'E'],
    'D': ['B', 'E'],  # Cycle: B -> D -> E -> C -> A -> B
    'E': ['C', 'D']
}

discovery, finish, predecessor, cycle_detected = dfs(graph)
print("Discovery Times:", discovery)
print("Finish Times:", finish)
print("Predecessors:", predecessor)
print("Cycle Detected:", cycle_detected)
```
